```python
# Завдання 1
import numpy as np
# Створення набору даних

# Використання arange для лінійних значень
linear_array = np.arange(0, 10, 2)  # Масив лінійних значень від 0 до 10 з кроком 2
print(f"Лінійний масив: {linear_array}")

# Використання random для генерації випадкових чисел
random_array = np.random.randint(1, 100, size=(3, 3))  # 3x3 матриця випадкових чисел від 1 до 100
print(f"Масив випадкових чисел:\n{random_array}")

# CRUD операції над масивами даних

# Операції з лінійним масивом (linear_array)
# Read (Читання)
print(f"Читання лінійного масиву: {linear_array}")

# Update (Оновлення)
linear_array[1] = 20  # Оновлення другого елемента масиву
print(f"Оновлений лінійний масив: {linear_array}")

# Delete (Видалення)
linear_array = np.delete(linear_array, 1)  # Видалення другого елемента масиву
print(f"Лінійний масив після видалення елемента: {linear_array}")

# Операції з масивом випадкових чисел (random_array)
# Read (Читання)
print(f"Читання масиву випадкових чисел:\n{random_array}")

# Update (Оновлення)
random_array[0, 0] = 99  # Оновлення елемента в першому рядку і першому стовпці
print(f"Оновлений масив випадкових чисел:\n{random_array}")

# Delete (Видалення)
random_array = np.delete(random_array, 1, axis=0)  # Видалення другого рядка
print(f"Масив випадкових чисел після видалення рядка:\n{random_array}")

random_array = np.delete(random_array, 0, axis=1)  # Видалення першого стовпця
print(f"Масив випадкових чисел після видалення стовпця:\n{random_array}")
```

```python
# Завдання 2
import pandas as pd
import numpy as np  # Для створення масиву, який буде використано у DataFrame

# Створення DataFrame різними способами

# 1. Створення DataFrame з масиву NumPy
numpy_array = np.random.randint(1, 100, size=(5, 3))  # Масив 5x3 з випадковими числами
df_from_array = pd.DataFrame(numpy_array, columns=['A', 'B', 'C'])
print("DataFrame з масиву NumPy:")
print(df_from_array)

# 2. Створення DataFrame з використанням словників
data_dict = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 32, 37, 29, 41],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']
}
df_from_dict = pd.DataFrame(data_dict)
print("\nDataFrame з використанням словників:")
print(df_from_dict)

# Маніпуляції з даними

# 1. Відображення перших рядків (head)
print("\nПерші 3 рядки DataFrame з масиву NumPy:")
print(df_from_array.head(3))  # Відображення перших 3 рядків

print("\nПерші 2 рядки DataFrame з використанням словників:")
print(df_from_dict.head(2))  # Відображення перших 2 рядків

# 2. Опис статистики (describe)
print("\nСтатистика DataFrame з масиву NumPy:")
print(df_from_array.describe())  # Опис статистики числових даних

print("\nСтатистика DataFrame з використанням словників:")
print(df_from_dict.describe(include='all'))  # Опис статистики для всіх типів даних

# 3. Індексація (iloc, loc)
# Індексація з використанням iloc (індексація за позиціями)
print("\nІндексація за позиціями (iloc) для DataFrame з масиву NumPy:")
print(df_from_array.iloc[1:3])  # Вибір другого та третього рядків

# Індексація з використанням loc (індексація за мітками)
print("\nІндексація за мітками (loc) для DataFrame з використанням словників:")
print(df_from_dict.loc[1:3, ['Name', 'City']])  # Вибір рядків від 1 до 3 та стовпців 'Name' і 'City'
```

```python
# Завдання 3
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Генерація даних для лінійної функції y = 2x + 3 з врахуванням помилок вимірювань
x = np.arange(0, 10, 0.5)
y_true = 2 * x + 3

# Додаємо помилку вимірювань (шум)
noise = np.random.normal(0, 1, size=x.shape)
y_observed = y_true + noise

# Побудова графіка лінійної функції та отриманих даних з помилками
plt.figure(figsize=(8, 5))
plt.plot(x, y_true, label='Лінійна функція y = 2x + 3', color='blue')
plt.scatter(x, y_observed, color='red', label='Дані з помилками')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Графік лінійної функції та отриманих даних з помилками')
plt.legend()
plt.grid(True)
plt.show()

# Обчислення метрик похибок (MAE та MSE)
mae = mean_absolute_error(y_true, y_observed)
mse = mean_squared_error(y_true, y_observed)

print(f"MAE (Mean Absolute Error): {mae}")
print(f"MSE (Mean Squared Error): {mse}")

# Запис результатів у CSV файл
results = {
    'x': x,
    'y_true': y_true,
    'y_observed': y_observed,
    'Error': y_true - y_observed
}

df_results = pd.DataFrame(results)

# Додаємо метрики похибок в кінці
metrics = {'MAE': [mae], 'MSE': [mse]}
df_metrics = pd.DataFrame(metrics)

# Запис даних у CSV файл
df_results.to_csv('results.csv', index=False)
df_metrics.to_csv('metrics.csv', index=False)

print("\nРезультати записано в файли 'results.csv' та 'metrics.csv'")
```
